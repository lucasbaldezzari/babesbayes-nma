# Clustering the data

```python
##importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os, requests
from numpy import pi
import seaborn as sns

from utils import getSubjecBasicStats

##train a PCA model using motion_direction, motion_coherence, reaction_time, estimate_angles, diff_angles
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
data = pd.read_csv("dataset\\data01_direction4priors.csv") ##open file
```

```python
subject_id = 6
data_subject = data.loc[data['subject_id'] == subject_id]
data_subject.describe()
```

#### Making ranges from reaction time column

I'will take the *reaction_time* and make a new column with *ranges* of reaction time using $0.1$ sec steps.

```python
ranges = np.arange(0,data_subject.loc[:,"reaction_time"].max(),0.1).round(2)
# print(ranges)
data_subject.loc[:,"reaction_time_range"] = pd.cut(data_subject.loc[:,"reaction_time"], bins=ranges, labels=False)+1

if data_subject.loc[:,'reaction_time_range'].isnull().sum() > 0: ##deleting rows with NaN values
    data_subject = data_subject.dropna(subset=['reaction_time_range'])

indexes = data_subject.loc[:,'reaction_time_range'].value_counts().sort_index().index.astype(int) #get the sorted index
x_labels = np.round(ranges[indexes],1) #filtering the range values using the sorted indexes
```

### Getting estimate angles

```python
from utils import getAngles
x = data_subject['estimate_x'].values
y = data_subject['estimate_y'].values
##using estimate_x and estimate_y to calculate the angle
data_subject.loc[:,'estimate_angles'] = np.round(getAngles(x, y), 3)
##limit the angles between 0 and 359 using %360
data_subject.loc[:,'estimate_angles'] = data_subject['estimate_angles'] % 360
```

```python
data_subject.loc[:,'diff_angles'] = data_subject['motion_direction'] - data_subject['estimate_angles']
data_subject.loc[:,'diff_angles'].describe()
```

## Preprossecing the data

```python
data.columns
```

```python
interest_columns = ["motion_direction","motion_coherence","reaction_time",'prior_std',
                    'response_arrow_start_angle',]

X = data_subject.loc[:,interest_columns] ##filtering the columns
scaler = StandardScaler() ##scaling the data
X_scaled = scaler.fit_transform(X) ##fitting the data

pca = PCA(n_components=len(interest_columns)) ##creating a PCA model
X_pca = pca.fit_transform(X_scaled) ##fitting the PCA
```

```python
##plotting the cumulative explained variance
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Explained Variance')
plt.show()
```

We can use 4 components

### Applying [DBSCAN](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html)

```python
#make a pipeline with standard scaler, PCA and DBSCAN
from sklearn.pipeline import make_pipeline
pipeline = make_pipeline(StandardScaler(),
                        #  PCA(n_components=4), 
                         DBSCAN(eps=0.5, min_samples=5))

pipeline.fit(X)

data_subject.loc[:,'cluster'] = pipeline.fit_predict(X_scaled)

##print the number of clusters
data_subject['cluster'].value_counts()
```

It seems the algorithm detect 42 clusters using the data before.

#### Plotting data for some cluster

I will make a scatter plot considering those trials for those cluster with more than 500 elements.

```python
##counting the number of clusters
cluster_counts = data_subject.loc[:,"cluster"].value_counts()
##getting the cluster with more than 1000 elements
filter = cluster_counts > 1000
principal_clusters = cluster_counts[filter].index
principal_clusters
```

```python
##use the principal_clusters to filter the data
data_principal_clusters = data_subject.loc[data_subject.loc[:,"cluster"].isin(principal_clusters)]
```

Scatter plot

```python
##scatter plot of "reaction_time" vs "motion_direction" with color coded clusters with hue = "cluster"
fig, axs = plt.subplots(1,2, figsize=(15,6))
# plt.figure(figsize=(10,6))
sns.scatterplot(data=data_principal_clusters, x="reaction_time", y="motion_direction", hue="motion_coherence",ax=axs[0])
sns.scatterplot(data=data_principal_clusters, x="reaction_time", y="motion_direction", hue="cluster",ax=axs[1])
plt.xlabel("Reaction Time")
plt.ylabel("Motion Direction")
axs[0].set_title("Reaction Time vs Motion Direction by motion_coherence")
axs[1].set_title("Reaction Time vs Motion Direction by clustering")
plt.show()
```
